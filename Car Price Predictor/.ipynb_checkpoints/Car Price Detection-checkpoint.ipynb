```python
def first_three(name):
    str = name.split(" ")
    if len(str) < 4 :
        return " ".join(str)
    return " ".join(str[0:3])
```

```python
# first_three("Hyundai Eon")
df['name_3'] = df['name'].apply(first_three)
```
