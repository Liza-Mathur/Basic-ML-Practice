```python
def first_three(name):
    str = name.split(" ")
    if len(str) < 4 :
        return " ".join(str)
    return " ".join(str[0:3])
```

```python
def remove_coma(text):
    flag = False
    num = 0
    for i in str(text):
        if i.isnumeric():
            flag=True
            num = num*10 + int(i)
    if flag:
        return num
    return None
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(X , y , test_size=0.25 , random_state=661)
```

```python
# X = pd.get_dummies(X , drop_first=True)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
from sklearn.metrics import r2_score , mean_absolute_error , mean_squared_error
```

```python
lr.fit(x_train , y_train)
lr_pred = lr.predict(x_test)
# print("MAE :" , mean_absolute_error(lr_pred , y_test))
# print("RMSE :" , np.sqrt(mean_squared_error(lr_pred , y_test)))
print("r2 score: " , r2_score(y_test , lr_pred))
```

```python
from sklearn.neighbors import KNeighborsRegressor
help(KNeighborsRegressor)
```

```python
knn = KNeighborsRegressor()
knn.fit(x_train , y_train)
knn_pred = knn.predict(x_test)
```

```python
r2_score(y_test , knn_pred)
```

```python
r = []
for k in range(10,100):
    kn = KNeighborsRegressor(n_neighbors=k)
    kn.fit(x_train , y_train)
    kn_pred = kn.predict(x_test)
    r.append(r2_score(y_test , kn_pred))
```

```python
plt.plot(range(10,100) , r)
```

```python
from sklearn.svm import SVR
svr = SVR(gamma=1 , kernel='poly')
svr.fit(x_train , y_train)
svr_pred = svr.predict(x_test)
print(r2_score(y_test , svr_pred))
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import Pipeline , make_pipeline
```

```python
X.columns
```

```python
ohe = OneHotEncoder()
ohe.fit(X[['name','company','fuel_type']])
```

```python
column_trans = make_column_transformer((OneHotEncoder(categories=ohe.categories_) , ['name','company','fuel_type']) , remainder = 'passthrough')
```

```python
lr = LinearRegression()
```

```python
pipe = make_pipeline(column_trans , lr)
```

```python
pipe.fit(x_train , y_train)
```

```python
pred = pipe.predict(x_test)
```

```python
print(r2_score(pred , y_test))
```
